```python
# required libraries 
import pandas as pd # data manipulation
import matplotlib.pyplot as plt # main visualization module: matplotlib's pyplot
import seaborn as sns # visualizations' top level. seaborn's style on top of matplotlib
sns.set()
```

```python
Gender = pd.read_excel("Gender.xlsx")
```

```python
Gender
```

```python
# Set the matplotlib color cycle using a seaborn palette
sns.set_palette('colorblind')
```

```python
# create a pie chart with the pie function from matplotlib's pyplot.
# Labels are added as the number of cars in each category in the form of percentages.
# Additional labels formatting using textprops. Note it is a dictionary.  

plt.figure(figsize = (8, 6))
plt.pie(Gender['Number of Respondents'], 
        labels = Gender['Gender'].values, 
        autopct = '%.2f%%',
        textprops = {'size' : 'x-large',
                     'fontweight' : 'bold', 
                     'rotation' : 'horizontal',
                     'color' : 'w'})
plt.legend(loc='lower left',fontsize='7')
plt.title('Gender of Respondents', fontsize = 14, fontweight = 'bold')
plt.savefig("Gender of Respondents.png", bbox_inches='tight') # you can export your chart as a picture
plt.show()
```
